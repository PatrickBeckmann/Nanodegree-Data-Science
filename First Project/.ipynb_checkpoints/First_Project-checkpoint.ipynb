# Data Preparation

## Boston 

```python
bos_obj = df_bos.select_dtypes(include=['object']).copy()
bos_obj.shape
```

```python
bos_obj = bos_obj.drop(list(bos_obj.filter(regex= "url")), axis = 1)
bos_obj.shape
```

```python
bos_obj.info()
```

```python
pd.set_option('display.max_columns', None)
bos_obj.head()
```

Convert to int where possible (e.g. Price, Dates, etc.)

```python
bos_obj['price'] = bos_obj['price'].str.replace('$', '')
bos_obj['price'] = bos_obj['price'].str.replace(',', '')
bos_obj['price'] = bos_obj['price'].astype(float)
```

```python
bos_obj['extra_people'] = bos_obj['extra_people'].str.replace('$', '')
bos_obj['extra_people'] = bos_obj['extra_people'].str.replace(',', '')
bos_obj['extra_people'] = bos_obj['extra_people'].astype(float)
```
